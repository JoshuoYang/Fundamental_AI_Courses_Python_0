# Lab 3.2.1
印出9x9乘法表  
1x1=1  
1x2=2  
...  
9x8=72  
9x9=81  


```python
for a in range(1, 10):
    for b in range(1, 10):
        print(a,'x',b,'=',a*b)
        # print("%dx%d=%d" % (a, b, a*b))
```

# Lab 3.2.2
印出9x9乘法表


```python
for a in range(1, 10):
    for b in range(1, 10):
        if b < 9:
            end = ' ' #end_str
        else:
            end = '\n'
        print("%dx%d=%2d" % (b, a, b*a), end=end) #%2d
    #print()
```
