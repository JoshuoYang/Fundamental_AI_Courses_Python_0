## Lab.1-1
輸入兩個數字以逗號+空白間隔並將其總和印出（遵從以下格式）。


```python
myInput = input("please input two numbers as a, b:\n").split(sep=', ')

a, b = map(int, myInput)

print("a is:%d" % (a))
print("b is:%d" % (b))

print("the sum of a, b is %d." % (a+b))
```

## Lab.1-2
依序輸入一串姓名、數字做為分數並以逗號+空白間隔，將數字調整後（開根號後乘10，並四捨五入到小數點後兩位）按照以下格式印出。


```python
import math
myInput = input("請輸入學生姓名、計算機概論成績:\n").split(sep=', ')

name, score = myInput[0], int(myInput[1])

print("%s經過調分以後的計算機概論成績是:%.2f" % (name, round(math.sqrt(score)*10, 2)))
```
